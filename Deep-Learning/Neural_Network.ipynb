```python
import numpy as np
```

## Simple dataset: predict XOR output

```python
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])
```

## Initialise weights

```python
np.random.seed(42)

W1 = np.random.randn(2,4)
b1 = np.zeros((1,4))
W2 = np.random.randn(4,1)
b2 = np.zeros((1,1))
```

## Activation Function

```python
def sigmoid(z):
  return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
  return sigmoid(z) * (1 - sigmoid(z))
```

## Training Loop

```python
learning_rate = 0.1
epochs = 1000

for epoch in range(epochs):

  # Forward Pass

  Z1 = np.dot(X, W1) + b1
  A1 = sigmoid(Z1)
  Z2 = np.dot(A1, W2) + b2
  A2 = sigmoid(Z2)

  # Loss

  loss = np.mean(y - A2) ** 2

  # Backward pass
  dA2 = -2 * (y - A2)
  dZ2 = dA2 * sigmoid_derivative(Z2)
  dW2 = np.dot(A1.T, dZ2)
  db2 = np.sum(dZ2, axis=0, keepdims=True)

  dA1 = np.dot(dZ2, W2.T)
  dZ1 = dA1 * sigmoid_derivative(Z1)
  dW1 = np.dot(X.T, dZ1)
  db1 = np.sum(dZ1, axis=0, keepdims=True)

  # Update weights
  W2 -= learning_rate * dW2
  b2 -= learning_rate * db2
  W1 -= learning_rate * dW1
  b1 -= learning_rate * db1

  # Print loss every 1000 epochs
  if epoch % 100 == 0:
      print(f"Epoch {epoch} — Loss: {loss:.4f}")
```

```python
# ── 5. Final predictions ──────────────────────────────────
print("\nFinal Predictions:")
print(np.round(A2))
print("Expected:", y)
```

# Explanation of backpropagation

1. dA  → how wrong was this layer's output
2. dZ  → how much did raw values contribute
3. dW  → how much did weights contribute
4. db  → how much did biases contribute

Line by line

- dA2 = -2 * (y - A2)

This is the first question — how wrong was the output?

y is actual answer

A2 is predicted answer

y - A2 is the error

-2 comes from derivative of MSE loss formula

If prediction was 0.9 but actual was 0 — error is huge. If prediction was 0.1 but actual was 0 — error is small. This line measures that.

- dZ2 = dA2 * sigmoid_derivative(Z2)

Before A2 was calculated, there was Z2 — the raw value before sigmoid. This line asks — how much did Z2 contribute to the output error?
sigmoid_derivative tells us how sensitive the sigmoid activation was at that point. If sigmoid was nearly flat there — small contribution. If steep — large contribution.

- dW2 = np.dot(A1.T, dZ2)

Now the key question — how much did each weight in W2 contribute to the error?
W2 connects hidden layer to output. This calculates the gradient for every weight in W2 — meaning how much should each weight change to reduce the error.
A1.T is transposed because of matrix dimension matching — same math, just shape alignment.

- db2 = np.sum(dZ2, axis=0, keepdims=True)

Same idea but for bias b2. How much should bias change? Sum across all training examples.

- dA1 = np.dot(dZ2, W2.T)

Now we go one layer deeper — back into the hidden layer.
This asks — how much did the hidden layer activations A1 contribute to the output error? We use W2 to carry the error signal backwards through the network.
This is the core of backpropagation — passing error signal backwards layer by layer.

- dZ1 = dA1 * sigmoid_derivative(Z1)

Same as dZ2 but for hidden layer. How much did raw hidden layer values Z1 contribute to error? Apply sigmoid derivative again.

- dW1 = np.dot(X.T, dZ1)

How much did each weight in W1 contribute to the error? X is the input — same logic as dW2 but one layer back.

- db1 = np.sum(dZ1, axis=0, keepdims=True)

Gradient for hidden layer biases. Same pattern as db2.
