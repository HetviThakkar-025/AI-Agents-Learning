```python
from langchain_groq import ChatGroq
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
# create tool

@tool
def multiply(a:int, b:int) -> int:
  """Multiply two numbers a and b"""
  return a*b
```

```python
print(multiply.invoke({"a":2,"b":3}))
```

```python
llm = ChatGroq(model="llama-3.1-8b-instant")
```

```python
# tool binding
llm_tools = llm.bind_tools([multiply])
```

```python
# tool calling

llm_tools.invoke("Hi how are you")
```

```python
query = HumanMessage("Can u multiply 2 with 1000")
```

```python
messages = [query]
```

```python
result = llm_tools.invoke(messages)
```

```python
messages.append(result)
```

```python
messages
```

```python
ai_result = result.tool_calls
```

```python
ai_result[0]
```

```python
tool_result = multiply.invoke(ai_result[0])
```

```python
messages.append(tool_result)
```
