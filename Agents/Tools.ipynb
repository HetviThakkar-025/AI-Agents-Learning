## 3. Custom tools

## Using @tool decorator

```python
from langchain_core.tools import tool
```

```python
def multipy(a,b):
  """multiply two numbers"""
  return a*b
```

```python
# type hinting

def multipy(a:int,b:int) -> int:
  """multiply two numbers"""
  return a*b
```

```python
# add tool decorator

@tool
def multipy(a:int,b:int) -> int:
  """multiply two numbers"""
  return a*b
```

```python
result = multipy.invoke({"a":2,"b":3})
print(result)
```

```python
print(multipy.name)
print(multipy.description)
print(multipy.args)
```

```python
print(multipy.args_schema.model_json_schema())
```

## Using StructuredTool & Pydantic

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
def multipy_fun(a:int,b:int) -> int:
  return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multipy_fun,
    name="multiply",
    description="multiply two numbers",
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({"a":2,"b":3})
print(result)
```

```python
print(multipy.name)
print(multipy.description)
print(multipy.args)
```

## 3. Using BaseTool Class

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to add")
    b: int = Field(required=True, description="The second number to add")
```

```python
class MultiplyTool(BaseTool):
  name: str = "multiply"
  description: str = "multiply two numbers"
  args_schema: Type[BaseModel] = MultiplyInput

  def _run(self, a:int, b:int) -> int:
    return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({"a":2,"b":3})
print(result)
```

```python
print(multipy.name)
print(multipy.description)
print(multipy.args)
```

## Toolkits

```python
from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolkit:
  def get_tools(self):
    return[add, multiply]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()
```

```python
for tool in tools:
  print(tool.name)
  print(tool.description)
  print(tool.args)
```
