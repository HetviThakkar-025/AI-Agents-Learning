```python
from transformers import pipeline
```

### Sentiment Analysis

```python
sentiment = pipeline("sentiment-analysis")
```

```python
print(sentiment("I love building AI agents"))
print(sentiment("Model is not good"))
```

### Named entity recognition

```python
ner = pipeline("ner", aggregation_strategy="simple")
```

```python
print(ner("My name is Hetvi and I study at LJ University in Gujarat"))
```

### Text generation

```python
generator = pipeline("text-generation", model="gpt2")
```

```python
result = generator("AI agents is", max_length=50)
print(result[0]['generated_text'])
```

### Zero shot classification

```python
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
print(classifier(
    "I love watching cricket and also love politics",
    candidate_labels=["coding", "cooking", "sports", "politics"]
))
```
